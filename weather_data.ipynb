```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd   # pandas is great for reading CSV files

# Load data from CSV
data = pd.read_csv("temperature_data.csv")

# Extract columns
days = data["Day"].to_numpy()       # Convert Day column to NumPy array
city_a = data["CityA"].to_numpy()   # Convert CityA column
city_b = data["CityB"].to_numpy()   # Convert CityB column

# Plot City A
plt.plot(days, city_a, marker='o', linestyle='-', color='b', label='City A')

# Plot City B
plt.plot(days, city_b, marker='s', linestyle='--', color='r', label='City B')

# Add labels, title, grid
plt.title("Weather Data from CSV")
plt.xlabel("Day")
plt.ylabel("Temperature (Â°C)")
plt.grid(True)
plt.legend()

plt.show()
```
