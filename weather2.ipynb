```python
import pandas as pd


data = {
    "Date": ["2023-01-01", "2023-01-02", "2023-01-03","2023-01-04", "2023-01-05"],
    "City": ["New York", "Los Angeles", "Chicago", "Houston", "Harare"],
    "Temperature": [30, 25, 20, 35, 28],
    "Humidity": [70, 65, 80, 75, 60],
    "Wind Speed": [10, 12, 8, 15, 5],
    "Weather Condition": ["Sunny", "Cloudy", "Rainy", "Sunny", "Windy"]
}
df = pd.DataFrame(data)
df["Date"] = pd.to_datetime(df["Date"])
df.set_index("Date", inplace=True)
df.to_csv("weather_data.csv", index=True)
print("Weather data has been saved to 'weather_data.csv'.")
```
