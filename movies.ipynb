```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('movie.csv')

print("Average rating by genre:")
print(data.groupby('Genre')['Rating'].mean())

print("\nTop 5 movies by rating:")
print(data.sort_values(by='Rating', ascending=False).head(5))

print("\nBest Rated Movies:")
print(data.loc[data["Rating"].idxmax()])

print("\nWorst Rated Movies:")
print(data.loc[data["Rating"].idxmin()])

print("\nCorrelation between Rating and Revenue:")
print(data['Rating'].corr(data['Revenue']))

plt.figure(figsize=(10, 6))
data.groupby('Genre')['Rating'].mean().plot(kind='bar', color='green', alpha=0.8)
plt.title('Average Movie Rating by Genre')
plt.xlabel('Genre')
plt.ylabel('Average Rating')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

plt.figure(figsize=(10, 6))
plt.scatter(data["Rating"], data["Revenue"], color='purple', alpha=0.5)
plt.title('Rating vs Revenue')
plt.xlabel('Rating')
plt.ylabel('Revenue')
plt.grid(True)
plt.tight_layout()
plt.show()
```
