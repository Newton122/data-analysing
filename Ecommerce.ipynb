```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("ecommerce.csv")

print("First look at data")
print(df.head())

total_sales = df["Amount"].sum()
print("\n Total Sales",total_sales)

category_sales = df.groupby("Category")["Amount"].sum()
print("Category sales", category_sales)


customer_sales = df.groupby("Customer")["Amount"].sum()
print("Customer sales", customer_sales)

daily_sales = df.groupby("Date")["Amount"].sum()
print("Daily sales", daily_sales)


category_sales.plot(kind = "bar", title = "Sales by Category")
plt.ylabel("Amount")
plt.show()

daily_sales.plot(kind = "line", marker = "o", title= "Daily Sales Trend")
plt.ylabel("Amount")
plt.show()


```
