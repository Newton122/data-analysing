```python
import seaborn as sns
import pandas as pd 
import matplotlib.pyplot as plt 

roll_no = [1,2,3,4,5,6,7,8]
marks = [10,15,20,30,10,25,45,20]
sample_pdf = pd.DataFrame({"Roll_no": roll_no, "Marks": marks})
sample_pdf.head()

sns.lineplot(data=sample_pdf, x="Roll_no", y="Marks", marker='o')
plt.title("Marks vs Roll Number")
```
